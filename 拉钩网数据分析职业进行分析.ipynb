```python
#批量读取文件, os模块获取文件列表
os.listdir('lagou_copy')
```

```python
filename_list = []
for filename in os.listdir('lagou_copy'):
#     print(filename)
    #只要csv文件
    if(filename.split('.')[-1] == 'csv'):
        print(filename)
        filename_list.append(filename)
```

```python
filename_list
```

```python
#批量读取,城市名作为key,数据作为value

lagou_dict = {}
for filename in filename_list:
    lagou_dict[filename.split('.')[0]] = pd.read_csv('lagou_copy/'+ filename)
print(lagou_dict)
```

```python
lagou_dict['zhengzhou']
```

## 批量删除表格里的第一列

```python
for file in lagou_dict.values():
    file.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
lagou_dict['zhengzhou']
```

## 统计每一个城市分别有多少条数据

```python
positions_num = pd.Series()
for key in lagou_dict:
    value = lagou_dict[key].shape[0]
    positions_num[key] = value
print(positions_num)
```

<!--  按照职位数排序 -->

```python
# 按照职位多少排序
positions_num.sort_values(ascending = False)
```

### 删除职位少于30条的数据


```python
#新建一个字典，存放职位数据大于等于30的城市数据
lagou_dict_30 = {}

for key in lagou_dict:
    if lagou_dict[key].shape[0] >= 30:
        lagou_dict_30[key] = lagou_dict[key]
print(lagou_dict_30)
```

```python
lagou_dict_30.keys()
```

### 批量统计这些表格的数据

```python
for key in lagou_dict_30:
    print(key)
    print(lagou_dict_30[key].isnull().sum())
```

### 对城市列进行填充（按照众数填充）: beijing 没有城市这一列，如果没有这一列的话，添加城市列兵填充

```python
# lagou_dict_30[城市名].城市.fillna(lagou_dict_30[城市名].城市.fillna(城市名))
```

```python
for key in lagou_dict_30:
    #有城市列直接填充
    if '城市' in lagou_dict_30[key].columns:
        lagou_dict_30[key].城市.fillna(lagou_dict_30[key].城市.mode()[0], inplace=True)
    #没有城市列，添加城市列并填充（暂时可能还没有发现北京没有城市列）
#     else:
#         lagou_dict[key]['城市'] = lagou_dict_30[key].城市.mode()[0]
```

```python
# 插在第一列后面， 此方法适合已经有该列了,因为insert第二个参数是原来的列名

col = lagou_dict_30['beijing'].城市

lagou_dict_30['beijing'] = lagou_dict_30['beijing'].drop('城市', axis=1)
lagou_dict_30['beijing'] 

lagou_dict_30['beijing'].insert(1, '城市', col)

lagou_dict_30['beijing']
```

# 进一步处理数据

   ## 去掉公司地点中的[ ]

```python
# 去掉公司地点中的[]
df['公司地点'] = df.公司地点.str.replace(']', '').str.replace('[', '')
```

## 把工位要求拆分成3列(数据有问题，有些工位要求的列没有薪资)

```python
t = df[~(df.工位要求.str.contains('k') | df.工位要求.str.contains('K'))] #这些工位要求列没有薪资,加上
t
```

```python
df.loc[t.index, '工位要求'] = ['8k-16k ', '6k-8k ', '6k-9k '] + df.loc[t.index, '工位要求']  # 拼接成功
df.iloc[1575:]  
```

```python
# 把工位要求拆分成三列
df[['薪资', '经验要求', '临时', '学历']] = df.工位要求.str.split(' ', expand = True)
```

```python
df
```

```python
# df.iloc[1570:]
```

### 去掉重复列

```python
#去掉重复列
df.drop(columns=['薪资待遇', '工位要求', '临时'], inplace = True) 
df
```

## 薪资列处理，拆分成最大值和最小值

```python
salary = df.薪资.str.replace('k', '000').str.replace('K', '000')
salary
```

```python
df[['最高薪资', '最低薪资']] = salary.str.split('-', expand = True)
```

```python
# 将薪资改成整数型
df.最高薪资 = df.最高薪资.astype('int')
df.最低薪资 = df.最低薪资.astype('int')
df
```

```python
df['平均薪资'] = (df.最高薪资 + df.最低薪资) // 2
```

### 地图

```python
from pyecharts.charts import Geo
from pyecharts import options as opts
```

```python
num = df.城市.value_counts()
num
```

```python
num.to_dict()
```

```python
num.to_dict().items()  #转换成功  格式：([('北京', 450), ('上海', 435)]
```

```python
geo = Geo()
# 地图类型
geo.add_schema(maptype='china')
data = num.to_dict().items()
# 传的数据类型
geo.add('数据分析岗位数量', data)
# 把默认的维度标签隐藏掉
geo.set_series_opts(label_opts=opts.LabelOpts(is_show=False))  
# 配置地图显示类型和标题
geo.set_global_opts(visualmap_opts=opts.VisualMapOpts(),title_opts = opts.TitleOpts(title='城市岗位分布'))
geo.render_notebook()
geo.render('city_positions.html')
```
